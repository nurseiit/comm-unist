#### Problem 4

Draw a negative rhombus with a given line length.

Example: If line=5, a result is as follows:

```
**** ****
***   ***
**     **
*       *

*       *
**     **
***   ***
**** ****
```

Problem: Draw a rhombus with line=7.

````python
def rhombus(line):
    single = lambda x: '*' * (line - x - 1) + ' ' * (x * 2 + 1) + '*' * (line - x - 1)
    result = []
    for i in range(line-1):
        result.append(single(i))
    print('\n'.join(result))
    print('\n')
    print('\n'.join(result[::-1]))

rhombus(7)
````

#### Problem 5

Let a list of animal names pets are given.

```
pets = ['dog', 'cat', 'parrot', 'squirrel', 'goldfish']
```

Define a dictionary pair which have key(name) and value(# of characters)

and print a dictionary and data as follows.

```
The length of dog : 3
The length of cat : 3
The length of parrot : 6
The length of squirrel : 8
The length of goldfish : 8
```

````python
pets = ['dog', 'cat', 'parrot', 'squirrel', 'goldfish']
dic = {}
for pet in pets:
    dic[pet] = len(pet)
    
print(dic)
for key in dic:
    print('The length of %s : %d' % (key, dic[key]))
````

Let $G_1=(V_1,E_1)$ and $G_2=(V_2,E_2)$ are defined by

* $V_1={3,4,5,6,7,8,9}$ and
* $E_1={(7,8),(8,9),(9,3),(3,5),(4,6),(3,7),(4,7),(3,4)}$
* $G_2=K_8$ where $K_8$: complete graph with $\vert V \vert=8$

$H_1=G_1 \cap G_2$
is defined as follows:

````python
G1=nx.Graph()
G1.add_nodes_from(range(3,10))
G1.add_edges_from([(7,8),(8,9),(9,3),(3,5),(4,6),(3,7),(4,7),(3,4)])
G2=nx.complete_graph(8)
# Intersection of G1nG2
V1,E1=set(G1.nodes()),set(G1.edges())
V2,E2=set(G2.nodes()),set(G2.edges())
H1=nx.Graph()
V3,E3=V1&V2, E1&E2
H1.add_nodes_from(V3)
H1.add_edges_from(E3)
````

#### Problem 1
Find $H_2=G_1 \cup  G_2$
and print network information (info and list of node,edge)

````python
H2 = nx.Graph()
V4,E4 = V1|V2, E1|E2
H2.add_nodes_from(V4)
H2.add_edges_from(E4)
print(nx.info(H2))
print(H2.nodes)
print(H2.edges)
````

#### Problem 2
Draw $H_2$ with the following conditions:

* Give $H_1$ and $H_2-H_1$ different colors, respectively.
    * Give nodes different colors.
    * Give edges different colors as well as styles.
    * Give labels


````python
V5 = V4-V3
H3 = nx.Graph()
H3.add_nodes_from(V5)

enormal = [(u, v) for (u, v) in H2.edges if not (u, v) in H1.edges]
edashed = H1.edges

pos = nx.spring_layout(H2)  # positions for all nodes

plt.figure(figsize=(10,10))

# nodes
nx.draw_networkx_nodes(H1, pos, node_size=700,
                      node_color='b')
nx.draw_networkx_nodes(H3, pos, node_size=700,
                      node_color='r')

# edges
nx.draw_networkx_edges(H2, pos, edgelist=enormal,
                       edge_color='b')
nx.draw_networkx_edges(H2, pos, edgelist=edashed,
                       edge_color='r', style='dashed')

# labels
nx.draw_networkx_labels(H2, pos)

plt.axis('off')
plt.show()
````

#### Problem 3
Load Zachary's karate club network and draw network with label and circular layout.

````python
K = nx.karate_club_graph()

pos = nx.circular_layout(K)

plt.figure(figsize=(6,6))
nx.draw(K, pos, node_size=300, node_color='r', with_labels=True)

plt.show()
````

````python
pos = nx.spring_layout(K, k=1)
sizes = [K.degree(v)*100 for v in K]
plt.figure(figsize=(8,8))
nx.draw(K, pos,
        node_color=range(len(K.nodes)), node_size=sizes, cmap=plt.cm.Reds,
        with_labels=True)

plt.show()
````

#### Problem 5

Draw adjacency matrix

Use cmap:Greys and interpolation="none"

````python
F = nx.to_numpy_matrix(K, dtype=np.bool)

plt.figure(1, figsize=(6, 6))

plt.imshow(F,
           cmap="Greys",
           interpolation="none")

plt.title('Zachary\'s club adjacency matrix', fontsize=12);
````

#### Problem 6
Draw a sorted node degree histogram

````python
sizes = reversed(sorted([(K.degree(v), v) for v in K]))
[weights, ids] = [list(t) for t in zip(*sizes)]

plt.figure(1, figsize=(10, 5))

x = range(0, len(ids))
plt.bar(x, height=weights)
plt.xticks(x, ids)
plt.xlabel('Node id')
plt.ylabel('Degree')
plt.title('Degree histogram')

plt.show()
````
